```python

def ask_questions(questions, correct_answers):
   user_answer = input(questions + "  " ).lower( ).strip( )

   if user_answer == correct_answers.lower( ).strip():
        print("Correct")
        return 1 
   else :
        print("Wrong ")
        print("Correct answer is !" , correct_answers)
        return 0 


def run_quiz():
    questions =  [ 
        "What is you name?" ,
        "Roll number ?" ,
        "Belong to ?"
 ]
    answers = [ 
        "Ahmad" , 
        "1197" ,
        "Mardan "
    ]

    score = 0 
    for i in range(len(questions)):
        print(f" questions{i+1}: ")
        score = score + ask_questions(questions[i], answers[i])


    percentage = score / len(questions) * 100

    print("Final Score " , score , "/" , len(questions))
    print("Percentage : " , percentage,"%")

run_quiz()
```

```python
students = [ 

    {"name" : "Ayaz" , "marks" : 80},
    {"name" : "muhammad" , "marks" : 40},
    {"name" : "Ahmad" , "marks" : 75},
    {"name" : "Ali" , "marks" : 30},
    {"name" : "khan" , "marks" : 65}
]

def calcul_aveg( students):
    total = 0
    
    for student in students :
        total = total + student["marks"]
    average = total / len(students)
    return average
def check_pass_fail(students):
    for student in students:
        if student["marks"] >= 50:
           print(student["name"] ,      student["marks"] ,  "passed")
        else:
            print(student["name"] ,      student["marks"], "failed")
average_marks = calcul_aveg(students)
print("Average Marks" , average_marks)

check_pass_fail(students)
        

```
